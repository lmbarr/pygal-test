```python
import pygal
import os
import webbrowser
print(webbrowser._browsers)
```

```python
bar_chart = pygal.Bar()
```

```python
bar_chart.title = 'Pet Population By Animal Per 100 households'
```

```python
bar_chart.add('Dogs', 41)
```

```python
bar_chart = pygal.Bar(height=360, width=640, explicit_size=True)
```

```python
bar_chart.add('Dogs', 41)
bar_chart.add('Cats', 20)
```

```python
bar_chart.title = 'Pet Population By Animal Per 100 households'
```

```python
bar_chart.render_to_file('pet_bar_chart.svg')
```

```python
def open_in_browser(file_to_open):
    file_path = os.path.realpath(file_to_open)
    file_url = 'file://' + file_path
    browser = webbrowser.get()
    browser.open(file_url)
```

```python
open_in_browser('pet_bar_chart.svg')
```

```python
horizontal_bar = pygal.HorizontalBar(height = 360, 
                                     width = 640, 
                                     explicit_size=True)

horizontal_bar.title = "Pet Population By Animal Per 100 households"
```

```python
horizontal_bar.add('Dogs',41)
horizontal_bar.add('Bird',12)
horizontal_bar.add('Horses', 7)
horizontal_bar.add('Exotic Pets',2)
horizontal_bar.add('No Pets',18)
```

```python
horizontal_bar.render_to_file('pet_horizontal_bar_chart.svg')
```

```python
open_in_browser('pet_horizontal_bar_chart.svg')
```
