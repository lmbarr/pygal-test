```python
import pygal 
import pandas as pd
```

```python
coal_df = pd.read_csv('datasets/production.csv')

coal_df.head()
```

```python
coal_df.tail()
```

```python
coal_df = coal_df[coal_df['Mineral'] == 'Coal']

coal_df.shape
```

```python
coal_df  = coal_df[['States', 'Quantity 2010-11',
                    'Quantity 2011-12', 'Quantity 2012-13', 
                    'Quantity 2013-14','Quantity 2014-15(P)']]
```

```python
coal_df
```

```python
coal_df = coal_df.drop([0])
```

```python
coal_df = coal_df.dropna()
```

```python
coal_df.columns=['States', '2010-11', '2011-12', '2012-13', '2013-14', '2014-15']
```

```python
coal_df = coal_df[coal_df['2014-15'] > 10000].reset_index()

coal_df
```

```python
coal_df = coal_df.drop(['index'], axis = 1)

coal_df
```

```python
coal_df.loc[:, '2010-11':'2014-15'] = coal_df.loc[:, '2010-11':'2014-15']\
                                             .div(1000)\
                                             .round()

coal_df
```

```python
from IPython.display import display, HTML

html_skeleton = """
<!DOCTYPE html>
<html>
  <head>
  <script type="text/javascript" 
          src="http://kozea.github.com/pygal.js/javascripts/svg.jquery.js">
  </script>
  <script type="text/javascript" 
          src="https://kozea.github.io/pygal.js/2.0.x/pygal-tooltips.min.js"">
  </script>
  </head>
  <body>
    <figure>
      {rendered_chart}
    </figure>
  </body>
</html>
"""

def display_chart(chart):
    rendered_chart = chart.render(is_unicode=True)
    plot_html = html_skeleton.format(rendered_chart=rendered_chart)
    display(HTML(plot_html))
```

```python
dot_chart = pygal.Dot(width = 640,
                      height = 400,
                      explicit_size = True,
                      truncate_legend = -1,
                      x_label_rotation = 30)
```

```python
dot_chart.title = 'Coal Production in Indian States'

dot_chart.x_labels = coal_df.columns[1:]
```

```python
dot_chart.value_formatter = lambda x: '{:.10g}M tons'.format(x)
```

```python
for index, row in coal_df.iterrows():
        dot_chart.add(row['States'], row['2010-11': '2014-15'])
```

```python
display_chart(dot_chart)
```
